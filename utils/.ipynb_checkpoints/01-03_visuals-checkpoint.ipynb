## Separate out waveforms and metadata according to plate length (2,4,6,8 in)

```python
from ae_functions import create_figure, plot_signal
import matplotlib.pyplot as plt
from ae_functions import get_signal_start_end

waves = data['waves']
location = data['location']
angle = data['angle']
length = data['length']
event = data['event']

import numpy as np
```

```python
w2in_topp = []
w4in_topp = []
w6in_topp = []
w8in_topp = []

w2in_side = []
w4in_side = []
w6in_side = []
w8in_side = []

w2in_tope = []
w4in_tope = []
w6in_tope = []
w8in_tope = []

for idx,wave in enumerate(waves):
    if length[idx] == '2in' and location[idx] == 'topp':
        w2in_topp.append(wave)
    if length[idx] == '4in' and location[idx] == 'topp':
        w4in_topp.append(wave)
    if length[idx] == '6in' and location[idx] == 'topp':
        w6in_topp.append(wave)
    if length[idx] == '8in' and location[idx] == 'topp':
        w8in_topp.append(wave)
         
    if length[idx] == '2in' and location[idx] == 'side':
        w2in_side.append(wave)
    if length[idx] == '4in' and location[idx] == 'side':
        w4in_side.append(wave)
    if length[idx] == '6in' and location[idx] == 'side':
        w6in_side.append(wave)
    if length[idx] == '8in' and location[idx] == 'side':
        w8in_side.append(wave)
        
    if length[idx] == '2in' and location[idx] == 'tope':
        w2in_tope.append(wave)
    if length[idx] == '4in' and location[idx] == 'tope':
        w4in_tope.append(wave)
    if length[idx] == '6in' and location[idx] == 'tope':
        w6in_tope.append(wave)
    if length[idx] == '8in' and location[idx] == 'tope':
        w8in_tope.append(wave)

print('Experiment 01')
print(f"# of 2 in top plate waves : {len(w2in_topp)}")
print(f"# of 4 in top plate waves : {len(w4in_topp)}")
print(f"# of 6 in top plate waves : {len(w6in_topp)}")
print(f"# of 8 in top plate waves : {len(w8in_topp)}\n")

print('Experiment 02')
print(f"# of 2 in side plate waves : {len(w2in_side)}")
print(f"# of 4 in side plate waves : {len(w4in_side)}")
print(f"# of 6 in side plate waves : {len(w6in_side)}")
print(f"# of 8 in side plate waves : {len(w8in_side)}\n")

print('Experiment 03')
print(f"# of 2 in top plate waves : {len(w2in_tope)}")
print(f"# of 4 in top plate waves : {len(w4in_tope)}")
print(f"# of 6 in top plate waves : {len(w6in_tope)}")
print(f"# of 8 in top plate waves : {len(w8in_tope)}\n")
```

## Compute and Plot Mean FFT (Entire Waveform)
The FFTs here are computed on the entire raw waveform (this is necessary to average them all and visualize, otherwise the ffts would have different lengths). Later in the code, the waveforms are chopped down to only what's contained between the start and end (what's visualized in the previous plotting of raw waveforms) before computed frequency domain features. 

```python
# Compute FFTs
low_pass = 0             # [Hz] ; low frequency cutoff
high_pass = 1000*10**3    # [Hz] ; high frequency cutoff\
dt = 10**-7              # [seconds] ; sample period / time between samples
fft_units = 1000

from ae_measure2 import fft

# 2 in_topp
fft2in_topp = []
for idx,wave in enumerate(w2in_topp):
    w,z = fft(dt, wave, low_pass, high_pass)
    fft2in_topp.append(z)
    
fft2in_topp = np.array(fft2in_topp)
mean_fft2in_topp= np.mean(fft2in_topp, axis=0) # average each column over all examples
std_fft2in_topp = np.std(fft2in_topp, axis=0)   # standard deviation

# 4 in_topp
fft4in_topp = []
for idx,wave in enumerate(w4in_topp):
    w,z = fft(dt, wave, low_pass, high_pass)
    fft4in_topp.append(z)
fft4in_topp = np.array(fft4in_topp)
mean_fft4in_topp= np.mean(fft4in_topp, axis=0) # average each column over all examples
std_fft4in_topp = np.std(fft4in_topp, axis=0)   # standard deviation

# 6 in_topp
fft6in_topp = []
for idx,wave in enumerate(w6in_topp):
    w,z = fft(dt, wave, low_pass, high_pass)
    fft6in_topp.append(z)
fft6in_topp = np.array(fft6in_topp)
mean_fft6in_topp= np.mean(fft6in_topp, axis=0) # average each column over all examples
std_fft6in_topp = np.std(fft6in_topp, axis=0)   # standard deviation

# 8 in_topp
fft8in_topp = []
for idx,wave in enumerate(w8in_topp):
    w,z = fft(dt, wave, low_pass, high_pass)
    fft8in_topp.append(z)
fft8in_topp = np.array(fft8in_topp)
mean_fft8in_topp= np.mean(fft8in_topp, axis=0) # average each column over all examples
std_fft8in_topp = np.std(fft8in_topp, axis=0)   # standard deviation

w = w/fft_units; # khz
```

## Get Sause Vector 
Note that in feature extraction, the waveform is chopped down to only contain what is in the start and end. The previously plotted ffts were based off the entire waveform, whereas the features extracted here are derivd from the waveform in between start and end markers.

```python
def plot_vs_length(title,b1,c1,b2,c2,b3,c3,a = ['2','4','6','8'],freq=False,pp=False):
    fig,spec2 = create_figure(title,columns=1,rows=1,width=5,height=5,default_font_size=18\
                         ,tick_font_size=20,legend_font_size=20,axes_font_size=20,
                         title_font_size=22)

    ax = fig.add_subplot(spec2[0,0])
    
    plt.scatter(a, b1,label='Exp 01 | Top')
    plt.scatter(a, b2,label='Exp 02 | Side')
    plt.scatter(a, b3,label='Exp 03 | Top Edge')
    
    plt.errorbar(a, b1, yerr=c1, fmt="o")
    plt.errorbar(a, b2, yerr=c2, fmt="o")
    plt.errorbar(a, b3, yerr=c3, fmt="o")

    if freq==True:
        plt.ylabel('Freq (kHz)')
    if pp == True:
        plt.ylabel('Norm. Area')
        
    plt.ylim([0.0,0.2])
    #plt.legend()
    plt.xlabel('Eff. Plate Len. (in)')
    plt.show()
    return

features = ['Avg Freq','Reverb Freq','Rise Freq','Peak Freq','Freq Centroid','WPF',
           'PP1 (0-150kHz)','PP2 (150-300kHz)','PP3 (300-450kHz)','PP4 (450-600kHz)',
            'PP5 (600-900kHz)','PP6 (900-1200kHz)']

for idx,feature in enumerate(features):
    
    if idx < 6:
        
        b1 = [mean_sv2in_topp[idx]/1000,mean_sv4in_topp[idx]/1000,mean_sv6in_topp[idx]/1000,mean_sv8in_topp[idx]/1000]
        c1 = [std_sv2in_topp[idx]/1000,std_sv4in_topp[idx]/1000,std_sv6in_topp[idx]/1000,std_sv8in_topp[idx]/1000]
        
        b2 = [mean_sv2in_side[idx]/1000,mean_sv4in_side[idx]/1000,mean_sv6in_side[idx]/1000,mean_sv8in_side[idx]/1000]
        c2 = [std_sv2in_side[idx]/1000,std_sv4in_side[idx]/1000,std_sv6in_side[idx]/1000,std_sv8in_side[idx]/1000]
        
        b3 = [mean_sv2in_tope[idx]/1000,mean_sv4in_tope[idx]/1000,mean_sv6in_tope[idx]/1000,mean_sv8in_tope[idx]/1000]
        c3 = [std_sv2in_tope[idx]/1000,std_sv4in_tope[idx]/1000,std_sv6in_tope[idx]/1000,std_sv8in_tope[idx]/1000]

        plot_vs_length(feature,b1,c1,b2,c2,b3,c3,freq=True)
        
    else:
        
        b1 = [mean_sv2in_topp[idx] ,mean_sv4in_topp[idx] ,mean_sv6in_topp[idx] ,mean_sv8in_topp[idx] ]
        c1 = [std_sv2in_topp[idx] ,std_sv4in_topp[idx] ,std_sv6in_topp[idx] ,std_sv8in_topp[idx] ]
        
        b2 = [mean_sv2in_side[idx] ,mean_sv4in_side[idx] ,mean_sv6in_side[idx] ,mean_sv8in_side[idx] ]
        c2 = [std_sv2in_side[idx] ,std_sv4in_side[idx] ,std_sv6in_side[idx] ,std_sv8in_side[idx] ]
        
        b3 = [mean_sv2in_tope[idx] ,mean_sv4in_tope[idx] ,mean_sv6in_tope[idx] ,mean_sv8in_tope[idx] ]
        c3 = [std_sv2in_tope[idx] ,std_sv4in_tope[idx] ,std_sv6in_tope[idx] ,std_sv8in_tope[idx] ]
        
        plot_vs_length(feature,b1,c1,b2,c2,b3,c3,pp=True)
        
```
