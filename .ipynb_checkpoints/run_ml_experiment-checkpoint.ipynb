## Visualize mislabeled indices wrt to train data

```python
from waves.ml.acoustic_emission_dataset import AcousticEmissionDataset

train = AcousticEmissionDataset(train_path,4096,10**-7,0,10**6,1000,26,'fft')

front_waves = []
front_fft = []
topedge_fft = []
topedge_waves = []
backcorner_fft = []
backcorner_waves = []

for idx,(x,y,index) in enumerate(train):
    if y[0] == 1:
        front_fft.append(x.numpy())
        front_waves.append(train.waves[idx].numpy())
    elif y[1] == 1:
        topedge_fft.append(x.numpy())
        topedge_waves.append(train.waves[idx].numpy())
    elif y[2] == 1:
        backcorner_fft.append(x.numpy())
        backcorner_waves.append(train.waves[idx].numpy())
        
mean_front_fft = np.mean(front_fft, axis=0) # average each column over all examples
mean_topedge_fft = np.mean(topedge_fft, axis=0) 
mean_backcorner_fft = np.mean(backcorner_fft, axis=0) 
mean_front_waves = np.mean(front_waves, axis=0) 
mean_topedge_waves = np.mean(topedge_waves, axis=0) 
mean_backcorner_waves = np.mean(backcorner_waves, axis=0) 
```

```python
print(f"Mislabeled Indices in Test DataLoader: {mislabeled}")
dataset_mislabeled = []
for idx in mislabeled:
    dataset_mislabeled.append(data_indices[idx])
print(f"Mislabeled Indices in Test Dataset: {dataset_mislabeled}")
```

```python
from waves.visuals import create_figure, plot_signal
import matplotlib.pyplot as plt
from waves.signal_processing import fft

for dataloader_idx in mislabeled:
    
    dataset_idx = data_indices[dataloader_idx]
    
    # Plot the averaged fft
    fig,spec2 = create_figure(f'Mean Training Set FFT wrt {dataset_idx} Mislabeled',columns=1,rows=1,width=16,height=8,default_font_size=18\
                             ,tick_font_size=25,legend_font_size=18,axes_font_size=28,
                             title_font_size=28)
    ax = fig.add_subplot(spec2[0,0])
    w,z = fft(10**-7, test.waves[dataset_idx], low_pass=0, high_pass=10**6)
    plt.plot(w,z,'--',linewidth=0.75,label=f'{test_label[dataloader_idx]} incorrectly label as {test_predicted[dataloader_idx]}')
    plt.plot(w,mean_front_fft,label='front')
    plt.plot(w,mean_topedge_fft,label='topedge')
    plt.plot(w,mean_backcorner_fft,label='backcorner')
    #plt.ylim([0,60])
    plt.xlim([0,10**6])
    plt.legend()
    plt.show()
    
    # Plot the averaged wavefrom
    fig,spec2 = create_figure(f'Mean Training Set Waveform wrt {dataset_idx} Mislabeled',columns=1,rows=1,width=16,height=8,default_font_size=18\
                             ,tick_font_size=25,legend_font_size=18,axes_font_size=28,
                             title_font_size=28)
    ax = fig.add_subplot(spec2[0,0])
    dt = 10**-7
    sig_length = 4096
    duration = sig_length*dt*10**6 # convert to us
    time = np.linspace(0,duration,sig_length) # discretization of signal time
    plt.plot(time, test.waves[dataset_idx],'--',linewidth=0.75,label=f'{test_label[dataloader_idx]} incorrectly label as {test_predicted[dataloader_idx]}')
    plt.plot(time,mean_front_waves,label='front')
    plt.plot(time,mean_topedge_waves,label='topedge')
    plt.plot(time,mean_backcorner_waves,label='backcorner')
    plt.legend()
    plt.show()
```
